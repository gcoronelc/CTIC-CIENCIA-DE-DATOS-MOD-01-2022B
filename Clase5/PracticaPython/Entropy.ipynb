**Shannon's Calculation**

Podemos ejecutar el cálculo de Shannon en python usando la biblioteca matemática que se muestra aquí:

```python
from math import log2

#Probabilidad de cara o sello al lanzar una fair coin
p = 0.5

information = -log2(p)

print('La informacion es: ', information)
```


image.png

Cuando cambiamos la probabilidad a 0.25, como en el caso de elegir el color correcto del disco, obtenemos este resultado:

```python
from math import log2

p = 0.25

information = -log2(p)

print('La informacion es: ', information)
```

Si bien parece que el aumento de información es lineal, ¿qué sucede cuando calculamos el lanzamiento de un solo dado o le pedimos a alguien que adivine un número entre 1 y 10? A continuacon los cálculos de información para una lista de probabilidades desde menos seguro (p = 0.1) hasta seguro (p = 1.0):

```python
from math import log2
import matplotlib.pyplot as plt
%matplotlib
%matplotlib inline

probabilities = [0.1, 0.125, 0.25, 0.5, 0.75, 1.0]

information = [-log2(p) for p in probabilities]

plt.plot(probabilities, information, marker='.')
plt.title('Probabilidad vs Informacion')
plt.xlabel('Probabilidad')
plt.ylabel('Informacion')

plt.show()

```

El gráfico muestra que a mayor incertidumbre, el crecimiento de la información es sublineal, no lineal.

**Calculo de la Entropia:**

Podemos usar la biblioteca Scipy para realizar el cálculo de entropía. La subbiblioteca de "estadísticas" de Scipy tiene cálculo de entropía que podemos usar. Aquí está el código para calcular la entropía para el escenario donde los cuatro discos tienen diferentes probabilidades:

image.png

```python
from scipy.stats import entropy

p = [0.5, 0.25, 0.125, 0.125]
e = entropy(p, base=2)
print('Entropia:', e)
```

El método de la entropía toma dos parametros: la lista de probabilidades y la base. Base=2 es la opción aquí ya que estamos usando un registro binario para el cálculo.

Obtenemos el mismo resultado que en la tabla anterior. Con un código mínimo, la biblioteca Scipy nos permite calcular rápidamente la entropía de Shannon.
